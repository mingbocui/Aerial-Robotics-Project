```python
%matplotlib inline

import evaluate as tp
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
df_log = tp.get_data_from_log()
```

```python
data = tp.evaluation_waypoint_navigation(df_log)

plt.figure()

plt.plot(data['time'], data['x_drone'], label="coordinate")
plt.plot(data['time'], [data['pwx'][0]] * len(data['x_drone']), label="waypoint 1")
plt.plot(data['time'], [data['pwx'][1]] * len(data['x_drone']), label="waypoint 2")
plt.plot(data['time'], [data['pwx'][2]] * len(data['x_drone']), label="waypoint 3")

plt.xlabel('time [s]')
plt.ylabel('x [m]')
plt.grid()
plt.legend()


plt.figure()
plt.plot(data['time'], data['y_drone'], label="coordinate")
plt.plot(data['time'], [data['pwy'][0]] * len(data['y_drone']), label="waypoint 1")
plt.plot(data['time'], [data['pwy'][1]] * len(data['y_drone']), label="waypoint 2")
plt.plot(data['time'], [data['pwy'][2]] * len(data['y_drone']), label="waypoint 3")

plt.xlabel('time [s]')
plt.ylabel('y [m]')
plt.grid()
plt.legend()


plt.figure()

plt.plot(data['time'], data['z_drone'], label="coordinate")
plt.plot(data['time'], [data['pwz'][0]] * len(data['z_drone']), label="waypoint 1")
plt.plot(data['time'], [data['pwz'][1]] * len(data['z_drone']), label="waypoint 2")
plt.plot(data['time'], [data['pwz'][2]] * len(data['z_drone']), label="waypoint 3")

plt.xlabel('time [s]')
plt.ylabel('z [m]')
plt.grid()
plt.legend()
```
