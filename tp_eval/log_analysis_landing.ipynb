```python
%matplotlib inline

import evaluate as tp
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
df_log = tp.get_data_from_log()
```

```python
data = tp.evaluation_landing(df_log)

plt.figure()

plt.plot(data['time'], data['x_drone'], label="drone")
plt.plot(data['time'], data['x_tr'], label="truck")
plt.plot(data['time'], data['x_tr_meas'], label="truck (measured)")
plt.plot(data['time'], data['x_tr_eval'], label="truck (estimated)")

plt.xlabel('time [s]')
plt.ylabel('x [m]')
plt.grid()
plt.legend()


plt.figure()

plt.plot(data['time'], data['y_drone'], label="drone")
plt.plot(data['time'], data['y_tr'], label="truck")
plt.plot(data['time'], data['y_tr_meas'], label="truck (measured)")
plt.plot(data['time'], data['y_tr_eval'], label="truck (estimated)")

plt.xlabel('time [s]')
plt.ylabel('y [m]')
plt.grid()
plt.legend()


plt.figure()

plt.plot(data['time'], data['z_drone'], label="drone")
plt.plot(data['time'], data['z_tr'], label="truck")
plt.plot(data['time'], data['z_tr_meas'], label="truck (measured)")
plt.plot(data['time'], data['z_tr_eval'], label="truck (estimated)")

plt.xlabel('time [s]')
plt.ylabel('z [m]')
plt.grid()
plt.legend()
```
